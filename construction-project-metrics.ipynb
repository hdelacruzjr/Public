```python
hours= pd.read_csv('/kaggle/input/project-timesheet/project_timesheet.csv')
```

```python
hours.head()
```

```python
hours = hours[(hours['job']== 'tower') & (hours['date'] != '1/1/23')]
```

```python
hours['total']= hours.sum(axis=1, numeric_only= True)
```

```python
hours.head()
```

```python
hours_by_floor= hours[['floor','total']]
```

```python
hours_by_floor = hours_by_floor.set_index('floor')
```

```python
by_floor= hours_by_floor.groupby('floor')
```

```python
total_by_floor= by_floor.sum()
```

```python
total_by_floor.head()
```

```python
ax= total_by_floor.plot.bar(color='darkblue')
ax.set_title('Hours distribution by floor - from 08/14/23',color= 'darkblue')
ax.set_xlabel('FLOORS')
ax.set_ylabel('HOURS')
```
